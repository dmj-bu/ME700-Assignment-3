## FEM Tutorial: 6-Node Triangle (`D2_nn6_tri`)
This notebook demonstrates how to use the 6-node quadratic triangle element for interpolation, visualization, and isoparametric mapping.

```python
import sys
sys.path.append('../src')

import discretization_demo_helper_fcns as demo
import discretization as di
import numpy as np
import matplotlib.pyplot as plt
```

###  Element Overview
The `D2_nn6_tri` element is a 6-node triangular element with quadratic interpolation. The nodes include three corner nodes and three mid-edge nodes.

### Evaluate Shape Functions at Reference Coordinates

```python
xi = np.array([1/3, 1/3])  # centroid of the reference triangle
N = di.shape_fcn("D2_nn6_tri", xi)
N
```

### Derivatives of Shape Functions

```python
dN_dxi = di.shape_fcn_derivative("D2_nn6_tri", xi)
dN_dxi
```

### Visualize Field Interpolation

```python
node_values = np.array([1, 2, 3, 4, 5, 6])
demo.plot_interpolate_field_natural_coords_single_element(
    "D2_nn6_tri_interpolation.png", "D2_nn6_tri", node_values
)
from IPython.display import Image
Image("D2_nn6_tri_interpolation.png")
```

### Isoparametric Mapping Example

```python
node_coords = np.array([[2, 1], [3, 3], [1, 2], [2.75, 2], [2, 2.25], [1.5, 1.75]])
node_values = np.array([1, 2, 3, 4, 5, 6])
demo.visualize_isoparametric_mapping_single_element(
    "D2_nn6_tri_mapping.png", "D2_nn6_tri", node_coords, node_values
)
Image("D2_nn6_tri_mapping.png")
```
